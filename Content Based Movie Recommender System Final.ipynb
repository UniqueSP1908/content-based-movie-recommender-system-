```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i["name"])
        return L
        
```

```python
 movies["genres"]= movies["genres"].apply(convert)
```

```python
movies.head()
```

```python
 movies["keywords"] = movies["keywords"].apply(convert)
```

```python
movies.head()
```

```python
movies["cast"][0]
```

```python
def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i["name"])
            counter+=1
        else:
             break
    return L
```

```python
movies["cast"]=movies["cast"].apply(convert3)
```

```python
movies.head()
```

```python
movies["crew"][0]
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            L.append(i["name"])
            break
    return L
```

```python
movies["crew"]=movies["crew"].apply(fetch_director)
```

```python
movies.head()
```

```python
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# now remove spaces from the sentences of the column "genres","keywords","cast","crew"
# we can use def function and apply a loop or we can use lambda function also
```

```python
def collapse(L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
        
        
```

```python
movies["cast"]=movies["cast"].apply(collapse)
```

```python
movies["crew"]=movies["crew"].apply(collapse)
```

```python
movies["genres"]=movies["genres"].apply(lambda x:[i.replace(" ","")for i in x]if x is not None else None)


# movies["genres"] = movies["genres"].apply(lambda x: [i.replace(" ", "") for i in x] if x is not None else None)
```

```python
movies["keywords"]=movies["keywords"].apply(lambda x:[i.replace(" ","")for i in x]if x is not None else None)
```

```python
movies.head()
```

```python
movies["tags"]= movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```
